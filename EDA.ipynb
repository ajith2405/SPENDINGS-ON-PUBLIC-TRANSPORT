```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv('/content/sample_data/Selected_variables.csv')

# Check for duplicates
duplicates = df[df.duplicated()]
print("Duplicates:")
print(duplicates)

# Check for missing values
missing_values = df.isnull().sum()
print("Missing Values:")
print(missing_values)

# Data Quality Report
data_quality_report = df.describe(include='all').transpose()
print("Data Quality Report:")
print(data_quality_report)

# Visualize outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.title("Boxplot of Variables")
plt.show()

# Run a correlation matrix
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()

# Univariate visualizations
for column in df.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[column], kde=True)
    plt.title(f"Distribution of {column}")
    plt.show()

# Bivariate visualizations
for column in df.columns:
    if column != "PUB_TRANS_SPENDING":
        plt.figure(figsize=(10, 6))
        sns.scatterplot(x=df[column], y=df["PUB_TRANS_SPENDING"])
        plt.title(f"{column} vs PUB_TRANS_SPENDING")
        plt.show()

# Additional bivariate visualizations (bar charts, line graphs, etc.)
# Example: Bar chart for categorical variable URBAN_STATUS
plt.figure(figsize=(10, 6))
sns.countplot(x="URBAN_STATUS", data=df)
plt.title("Count of each URBAN_STATUS category")
plt.show()
```
